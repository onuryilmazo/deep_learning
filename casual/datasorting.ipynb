```python
from tqdm import tqdm
import random
# Function to generate random birthdates
def generate_random_birthdate():
    day = random.randint(1, 28)  # Set to 28 to account for February
    month = random.randint(1, 12)
    year = random.randint(1950, 2023)
    
    if month == 4 or month == 6 or month == 9 or month == 11:
        day = random.randint(1, 30)  # April, June, September, and November have 30 days

    if month == 2:
        day = random.randint(1, 29)  # February has 29 days for leap years
    
    birthdate = f"{day:02d}/{month:02d}/{year}"
    return birthdate

# Generate 87 million random birthdates
total_records = 87000000

# Create a tqdm progress bar
with tqdm(total=total_records, unit=" dates") as pbar:
    birthdates = []
    for _ in range(total_records):
        date = generate_random_birthdate()
        birthdates.append(date)
        pbar.update(1)  # Update the tqdm progress bar

# Create a Pandas DataFrame and save it to a CSV file
df = pd.DataFrame({'Birthdate': birthdates})
df.to_csv('random_birthdates.csv', index=False)

print("CSV file created: random_birthdates.csv")
```

```python
data = pd.read_csv('random_birthdates.csv')
```

```python
data.head()
```

```python
# Veriyi küçük parçalara bölmek için adım
chunk_size = 1000000  # Parça boyutunu istediğiniz gibi ayarlayın

# Verileri daha küçük gruplara bölmek ve sıralamak
sorted_chunks = []

# tqdm kullanarak işlemi takip etmek için bir döngü başlatın
for chunk_start in tqdm(range(0, len(data), chunk_size), desc='Processing Chunks'):
    chunk = data.iloc[chunk_start:chunk_start + chunk_size]
    # Geçersiz tarihleri ve eksik bilgileri atla
    chunk['Birthdate'] = pd.to_datetime(chunk['Birthdate'], format='%d/%m/%Y', errors='coerce')
    chunk = chunk.dropna(subset=['Birthdate'])
    chunk = chunk.sort_values(by='Birthdate')
    sorted_chunks.append(chunk)

# Tüm parçaları birleştirin ve sıralı veriyi yeni bir CSV dosyasına kaydedin
sorted_data = pd.concat(sorted_chunks)
sorted_data.to_csv('sirali_veriler.csv', index=False)
```
