"Model pipeline" veya kısaca "pipeline," makine öğrenme (ML) veya veri analitiği projelerinde kullanılan bir kavramdır. Model pipeline, veri işleme, model oluşturma ve sonuçların değerlendirilmesi aşamalarını bir araya getiren ve bu aşamaların düzenlenmesini kolaylaştıran bir yapısı ifade eder. Bu yapı, bir dizi işlemi sıralı olarak ve otomatik olarak gerçekleştiren bir akıştan oluşur.

Model pipeline'lar, aşağıdaki ana bileşenleri içerebilir:

Veri Ön İşleme (Data Preprocessing): Veri analizi ve model eğitimi öncesinde verinin temizlenmesi, dönüştürülmesi ve hazırlanmasını içerir. Bu aşamada eksik verilerin doldurulması, outlier'ların işlenmesi, kategorik verilerin sayısal verilere dönüştürülmesi gibi işlemler yapılır.

Özellik Mühendisliği (Feature Engineering): Veri setindeki özelliklerin daha anlamlı hale getirilmesi ve yeni özelliklerin oluşturulması sürecini ifade eder. Bu aşamada özellik seçimi ve özellik mühendisliği teknikleri kullanılır.

Model Seçimi (Model Selection): Veri setinin özelliklerine ve hedefe bağlı olarak uygun bir makine öğrenme modeli veya algoritma seçilir.

Model Eğitimi (Model Training): Seçilen model, eğitim verileri kullanılarak eğitilir.

Model Değerlendirmesi (Model Evaluation): Modelin performansı test verileri veya doğrulama verileri kullanılarak değerlendirilir. Doğruluk, hassasiyet, geri çağırma gibi metrikler kullanılarak değerlendirme yapılır.

Sonuçların Sunumu (Result Presentation): Modelin sonuçları kullanıcılar veya ilgili kişiler için anlaşılır bir şekilde sunulur.

Model Deploy (Model Dağıtım): Eğer model başarıyla değerlendirildiyse, canlı bir sistemde kullanılmak üzere deploy edilir.

Model pipeline, bu aşamaları sırayla ve düzenli bir şekilde yönetmeyi kolaylaştırır. Bu, özellikle büyük ve karmaşık ML projelerinde, birden çok veri kaynağı ve çok sayıda işlem içeren projelerde önemlidir. Ayrıca, pipeline'lar işlemleri otomatikleştirerek hata yapma olasılığını azaltabilir ve projelerin tekrarlanabilirliğini artırabilir.

Şimdi artık moedelimizi oluşturmaya geçebiliriz. Tensorflow ve Keras kullanacağız.

Bu kod, temel bir çok katmanlı sinir ağı (MLP) modeli oluşturur ve eğitir

```python
# Hedef değişkeni çıkarın
y_train = concat_data['Transported']
X_train = concat_data.drop('Transported', axis=1)
```

```python
y_train
```

```python
X_train
```

```python
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model = keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
    layers.Dense(64, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
```

```python
#modeli derleme
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
#model eğitimi
model.fit(X_train, y_train, epochs=30, batch_size=32)
```

```python
y_pred = model.predict(test_data)
```

```python
test_id = pd.read_csv('test.csv')
```

```python
predictions=pd.DataFrame({'PassengerID': test_id['PassengerId'],'Transported':y_pred.flatten() >0.5})
```

```python
predictions.to_csv('predictions.csv', index=False)
```
